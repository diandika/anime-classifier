```python
import os
import pandas as pd
import random
import shutil
from PIL import Image
```

```python
df = pd.read_csv("data_file/AnimeList.csv", sep=',')
df = df[df["image_url"].notna() & df["genre"].notna()]
df.head(15)
```

```python
anime_list= []
image_base_file = "image_file"
image_format = ".jpg"
corrupted_count = 0

for idx, row in df.iterrows():
    try:
        image_path = "%s/%s%s" % (image_base_file, row["anime_id"], image_format)
        Image.open(image_path)
        anime_dictionary = {
            "anime_id": row["anime_id"],
            "title": row["title"],
            "image_path": image_path,
            "genre": row["genre"]
        }
        anime_list.append(anime_dictionary)
    except:
        corrupted_count = corrupted_count + 1
        print(corrupted_count, " found corrupted file, ", image_base_file, row["anime_id"], image_format)
        pass
print(len(anime_list))
```

```python
H = ['Hentai', 'Yaoi', 'Yuri']
E = 'Ecchi'
H_anime = []
E_anime = []
non_EnH_anime = []
for row in anime_list:
    isH = False
    genre = row["genre"].split(', ')
    
    for item in H:
        if item in genre:
            isH = True
            break
    if isH:
        H_anime.append(row)
    else:
        if E in genre:
            E_anime.append(row)
        else:
            non_EnH_anime.append(row)
            
print("Hentai count: ", len(H_anime))
print("Ecchi count: ", len(E_anime))
print("Safe count", len(non_EnH_anime))
```

```python
def make_directory(path):
    try:
        os.mkdir(path)
    except:
        print("Path already exist")
        pass
```

```python
TYPE_CLASSIFICATION_DIR = "h_dir"
make_directory(TYPE_CLASSIFICATION_DIR)
TRAIN_DIR = "train"
TEST_DIR = "test"

train_path = os.path.join(TYPE_CLASSIFICATION_DIR, TRAIN_DIR)
make_directory(train_path)
test_path = os.path.join(TYPE_CLASSIFICATION_DIR, TEST_DIR)
make_directory(test_path)

h_train_path = os.path.join(train_path, 'hentai')
noh_train_path = os.path.join(train_path, 'non-hentai')
h_test_path = os.path.join(test_path, 'hentai')
noh_test_path = os.path.join(test_path, 'non-hentai')

make_directory(h_train_path)
make_directory(noh_train_path)
make_directory(h_test_path)
make_directory(noh_test_path)
```
